```python
import requests
import re
import os
import traceback
from bs4 import BeautifulSoup
from fake_useragent import UserAgent
from urllib.parse import quote
from urllib.request import Request, urlopen
```

```python
CITY_LIST = ["shanghai", "beijing"]
URL = "http://www.dianping.com"
```

```python
class crawler(object):
    
    def __init__(self):
        self.refer_url = "http://www.dianping.com/{}"
        self.food_url_format = "http://www.dianping.com/{}/ch10"

    def getHtmlText(self, url, refer_url=None):
        if refer_url is None: 
            refer_url = self.refer_url
        headers = {
                    "User-Agent": UserAgent(verify_ssl=False).random,
                    "Host":"www.dianping.com",
                    "Referer": refer_url
                   }
        print("from {} \n to {}".format(refer_url, url))
        try:
            r=requests.get(url,headers=headers)
            r.raise_for_status()
            self.refer_url = url
            if (r.encoding != "UTF-8"):
                r.encoding = r.apparent_encoding
            return r.text
        except:
            return ""
    
    def parseHtmlText(self, url, refer_url=None): 
        html= self.getHtmlText(url, refer_url)
        soup = BeautifulSoup(html, "html.parser")
        return soup
    
    def start(self, city_list):
        for c in city_list:
            root_url = self.refer_url.format(c)
            food_url = self.food_url_format.format(c)
            soup = self.parseHtmlText(food_url, root_url)
#             print(soup)
            area_urls = [area.attrs["href"] for area in soup.find("div", id="bussi-nav").find_all('a')]
#             print(area_urls)
            for a in area_urls:
                sub_soup = self.parseHtmlText(a)
                next_page_button = sub_soup.find("a", {"class": "next", "title": "下一页"})
                while(next_page_button is not None):
                    shops = sub_soup.find_all("div", class_="tit")
                    for shop in shops:
                        title = shop.find("a", {"data-click-name": "shop_title_click"}).attrs["title"]
                        #TODO: change print to other stuff
                        print(title)
                    next_link = next_page_button.attrs["href"]
                    sub_soup = self.parseHtmlText(next_link)
                    next_page_button = sub_soup.find("a", {"class": "next", "title": "下一页"})
                
                
                
            
    
```
